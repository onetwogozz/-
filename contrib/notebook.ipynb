```python
def extract_tree(node):
    if node is None:
        return None

    result = {
        "smiles": getattr(node, "smiles", None),
        "children": []
    }

    # ⭐ AIZynthFinder 常见真实字段（重点）
    children = (
        getattr(node, "children_nodes", None)
        or getattr(node, "child_nodes", None)
        or getattr(node, "children", None)
        or []
    )

    # ⚠️ 如果还是空，说明结构不是 tree，而是 graph
    if not children:
        return result

    for c in children:
        result["children"].append(extract_tree(c))

    return result
```
